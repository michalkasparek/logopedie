```python
def replace_sixth_space(string):
    count = 0
    new_string = ""
    for char in string:
        if char == " ":
            count += 1
            if count % 6 == 0:
                new_string += "\n"
            else:
                new_string += char
        else:
            new_string += char
    return new_string
```

## slova bez samohlásek

```python
def bez_samohlasek(x):
    if isinstance(x, str):
        counter = 0
        counter_max = 0
        vowels = ['a', 'á', 'e', 'é', 'ě', 'i', 'í', 'o', 'ó', 'u', 'ú', 'ů', 'y', 'ý', 'ö', '-', "'", 'ä', '/', 'ü', '.']
        for l in x:
            if l.lower() not in vowels:
                counter += 1
            else:
                counter = 0
            if counter > counter_max:
                counter_max = counter
        return counter_max
    else:
        return None
```

```python
df['bez_samohlasek'] = df['lemma'].apply(lambda x: bez_samohlasek(x))
```

```python
df[['lemma','bez_samohlasek']].sort_values(by='bez_samohlasek', ascending=False).head(60)
```

## nejvíckrát konkrétní hláska

```python
def kolikrat(hlaska):
    pracovni = df.copy()
    pracovni['kolikrat'] = pracovni['lemma'].apply(lambda x: str(x).count(hlaska))
    return pracovni[['lemma','kolikrat','frekvence']].sort_values(by=['kolikrat','frekvence'],ascending=False).head(20)
```
