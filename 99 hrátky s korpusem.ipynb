```python
import os
import pandas as pd
```

```python
df = pd.read_csv(os.path.join("data","syn2000_lemma_utf8.tsv"), sep='\t', header=None)
```

```python
df = df.rename(columns={0: 'rank', 1: 'lemma', 2: 'frekvence', 3: 'celková přepočítaná frekvence', 4: 'beletrie', 5: 'oborová', 6: 'publicistika', 7: 'charakteristika'})
```

## …knižně

```python
beletrie = (df.groupby('lemma')['beletrie'].max() / df.groupby('lemma')['publicistika'].max())
beletrie.sort_values()
```

```python
publicistika_series = (df.groupby('lemma')['publicistika'].max() / df.groupby('lemma')['oborová'].max())
publicistika = publicistika_series[(publicistika_series.isnull()) | (publicistika_series >= 1)].index.to_list()
publicistika = [p for p in publicistika if p[0] == p.lower()[0]]
', '.join(publicistika)
```

```python
beletrie_series = (df.groupby('lemma')['beletrie'].max() / df.groupby('lemma')['publicistika'].max())
beletrie = beletrie_series[(beletrie_series.isnull()) | (beletrie_series >= 1)].index.to_list()
beletrie = [b for b in beletrie if b[0] == b.lower()[0]]
', '.join(beletrie)
```

```python
beletrie_series2 = beletrie_series[(beletrie_series < 1000000) & (beletrie_series.index.str[0] == beletrie_series.index.str.lower().str[0])]
```

```python
beletrie_top = beletrie_series2.sort_values(ascending=False).head(250).index.to_list()
beletrie_top.sort()
", ".join(beletrie_top)
```

```python
with open("slova_z_beletrie_co_nejsou_v_publicistice.txt", "w+", encoding="utf-8") as slova:
    slova.write("\n".join(beletrie_top))
```
