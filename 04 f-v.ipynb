```python
import os
import pandas as pd
```

```python
df = pd.read_csv(os.path.join("data","syn2000_lemma_utf8.tsv"), sep='\t', header=None)
df = df.rename(columns={0: 'rank', 1: 'lemma', 2: 'frekvence', 3: 'celková přepočítaná frekvence', 4: 'beletrie', 5: 'oborová', 6: 'publicistika', 7: 'charakteristika'})
```

```python
print(', '.join(df[df['lemma'].str.lower().str.contains('v[áaiíéěůuyýoó]',na=False) & ~df['lemma'].str.lower().str.contains(r'[rř]',na=False)].sort_values(by='beletrie',ascending=False)['lemma'].head(1000).to_list()))
```

```python
print(', '.join(df[df['lemma'].str.lower().str.contains('f',na=False) & ~df['lemma'].str.lower().str.contains(r'[rř]',na=False)].sort_values(by='beletrie',ascending=False)['lemma'].head(1000).to_list()))
```

```python
print(', '.join(df[(df['lemma'].str.lower().str.contains('f',na=False)) & (~df['lemma'].str.lower().str.contains('r',na=False))]['lemma'].head(1000).to_list()))
```

```python
print(', '.join(df[(df['lemma'].str.lower().str.contains('va',na=False)) & (~df['lemma'].str.lower().str.contains('r',na=False))]['lemma'].head(1000).to_list()))
```
